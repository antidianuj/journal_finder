Modification of Typos in Bibiliography

```python
import re
lines = []
with open('ref.txt', encoding="utf8") as f:
    lines = f.readlines()

count = 0

journal=[]
for line in lines:
  #  count += 1
  #  print(f'line {count}: {line}') 

    if "journal" in line:
      try:
        journal.append(re.search('journal={(.*?)},', line).group(1))
        # print("***")
      except AttributeError:
        print("---")
```

```python
import pandas
from collections import Counter
letter_counts = Counter(map(str.lower, journal))
df = pandas.DataFrame.from_dict(letter_counts, orient='index')
df.plot(figsize=(70,7),kind='bar',grid=True)
```

```python
ten_letter_counts=letter_counts.most_common(30)
def Convert(tup, di):
    for a, b in tup:
        di.setdefault(a, []).append(b)
    return di

dictionary = {}
di=Convert(ten_letter_counts, dictionary)

letter_counts = Counter(journal)
df = pandas.DataFrame.from_dict(di, orient='index')
df.plot(figsize=(20,7),kind='bar',width=0.5,grid=True)
```
